```python
#Change symbolics to numerics 
import numpy as np
adultData['sex'] = np.where(adultData['sex'] == ' Male', 1, 0)
adultData['race'] = np.where(adultData['race'] != ' White',1, 0)
adultData['Probability'] = np.where(adultData['Probability'] == ' <=50K', 0, 1)
#dataset['sex'] = dataset['sex'].map({'Male':0, 'female':1})
## Discretize age
adultData['age'] = np.where(adultData['age'] >= 70, 70, adultData['age'])
adultData['age'] = np.where((adultData['age'] >= 60 ) & (adultData['age'] < 70), 60, adultData['age'])
adultData['age'] = np.where((adultData['age'] >= 50 ) & (adultData['age'] < 60), 50, adultData['age'])
adultData['age'] = np.where((adultData['age'] >= 40 ) & (adultData['age'] < 50), 40, adultData['age'])
adultData['age'] = np.where((adultData['age'] >= 30 ) & (adultData['age'] < 40), 30, adultData['age'])
adultData['age'] = np.where((adultData['age'] >= 20 ) & (adultData['age'] < 30), 20, adultData['age'])
adultData['age'] = np.where((adultData['age'] >= 10 ) & (adultData['age'] < 10), 10, adultData['age'])
adultData['age'] = np.where(adultData['age'] < 10, 0, adultData['age'])
```

```python
## Replace '?' marks with 'NA'
adultData = adultData.astype(str)
adultData = adultData.apply(lambda x: x.str.strip())
adultData = adultData.replace('?', 'NA')
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

adultData.workclass = label_encoder.fit_transform(adultData.workclass)
adultData.occupation = label_encoder.fit_transform(adultData.occupation)
adultData.relationship = label_encoder.fit_transform(adultData.relationship)
adultData.education = label_encoder.fit_transform(adultData.education)

adultData = adultData.rename(columns={'native-country': 'native_country'})
adultData = adultData.rename(columns={'marital-status': 'marital_status'})

adultData.native_country = label_encoder.fit_transform(adultData.native_country)
adultData.marital_status = label_encoder.fit_transform(adultData.marital_status)
```

```python
adultData
```

```python
features_df = adultData.iloc[:, :14]
features_array = features_df.to_numpy()
features_array = features_array.astype(int)
```

```python
features_array
```

```python
target_col = adultData['Probability']
target_array = target_col.to_numpy()
target_array = target_array.astype(int)
```

```python
target_array
```

```python
##Extract meta features related to simple, statistical, and information theoritic 

from pymfe.mfe import MFE 
mfe = MFE(features=["attr_to_inst","freq_class","inst_to_attr","can_cor","cor","cov","nr_disc","kurtosis","nr_norm","sd_ratio","skewness","w_lambda","attr_ent","class_ent","joint_ent","mut_inf"])
mfe.fit(features_array, target_array)
ft = mfe.extract()
```
