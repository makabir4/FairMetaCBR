```python
#Change symbolics to numerics 
import numpy as np
adultData['sex'] = np.where(adultData['sex'] == ' Male', 1, 0)
adultData['race'] = np.where(adultData['race'] != ' White',1, 0)
adultData['Probability'] = np.where(adultData['Probability'] == ' <=50K', 0, 1)
#dataset['sex'] = dataset['sex'].map({'Male':0, 'female':1})
## Discretize age
adultData['age'] = np.where(adultData['age'] >= 70, 70, adultData['age'])
adultData['age'] = np.where((adultData['age'] >= 60 ) & (adultData['age'] < 70), 60, adultData['age'])
adultData['age'] = np.where((adultData['age'] >= 50 ) & (adultData['age'] < 60), 50, adultData['age'])
adultData['age'] = np.where((adultData['age'] >= 40 ) & (adultData['age'] < 50), 40, adultData['age'])
adultData['age'] = np.where((adultData['age'] >= 30 ) & (adultData['age'] < 40), 30, adultData['age'])
adultData['age'] = np.where((adultData['age'] >= 20 ) & (adultData['age'] < 30), 20, adultData['age'])
adultData['age'] = np.where((adultData['age'] >= 10 ) & (adultData['age'] < 10), 10, adultData['age'])
adultData['age'] = np.where(adultData['age'] < 10, 0, adultData['age'])
```

```python
## Replace '?' marks with 'NA'
adultData = adultData.astype(str)
adultData = adultData.apply(lambda x: x.str.strip())
adultData = adultData.replace('?', 'NA')
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
label_encoder.fit(adultData['workclass'])

adultData['workclass_category_encoded'] = label_encoder.transform(adultData['workclass'])
print(adultData[['workclass', 'workclass_category_encoded']])
```

```python
label_encoder.fit(adultData['occupation'])
adultData['occupation_category_encoded'] = label_encoder.transform(adultData['occupation'])
print(adultData[['occupation', 'occupation_category_encoded']])
```

```python
label_encoder.fit(adultData['relationship'])
adultData['relationship_category_encoded'] = label_encoder.transform(adultData['relationship'])
print(adultData[['relationship', 'relationship_category_encoded']])
```

```python
label_encoder.fit(adultData['education'])
adultData['education_category_encoded'] = label_encoder.transform(adultData['education'])
print(adultData[['education', 'education_category_encoded']])
```

```python
adultData = adultData.rename(columns={'native-country': 'native_country'})
label_encoder.fit(adultData['native_country'])
adultData['native_country_category_encoded'] = label_encoder.transform(adultData['native_country'])
print(adultData[['native_country', 'native_country_category_encoded']])
```

```python
adultData = adultData.rename(columns={'marital-status': 'marital_status'})
label_encoder.fit(adultData['marital_status'])
adultData['marital_status_category_encoded'] = label_encoder.transform(adultData['marital_status'])
print(adultData[['marital_status', 'marital_status_category_encoded']])
```
